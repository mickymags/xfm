# Precision and Recall

```python
def stat_reporter(img, desc):
  reference_re = ref_clip.rename('reference')
  img_re = img.rename(desc)

  combo = reference_re.addBands(img_re)

  sample = combo.sample(
      region=aoi,
      scale=4,
      projection='EPSG:32603',
      numPixels=1e13
  )

  tp = 0
  tn = 0
  fp = 0
  fn = 0

  sample_size = sample.size().getInfo()
  sample_info = sample.toList(sample_size).getInfo()

  for j in range(sample_size):
    sample_of_int = sample_info[j]['properties']
    prop = sample_of_int[desc]
    reference_of_int = sample_of_int['reference']

    if prop == 1 and reference_of_int == 1:
      tp += 1
    elif prop == 1 and reference_of_int == 0:
      fp += 1
    elif prop == 0 and reference_of_int == 1:
      fn += 1
    elif prop == 0 and reference_of_int == 0:
      tn += 1

  prec = tp / (tp + fp)
  rec = tp / (tp + fn)
  acc = (tp + tn) / (tp + fp + tn + fn)


  ##### Intersection over Union #####
  intersec = img.And(ref_clip)
  union = img.Or(ref_clip)

  inter_mask = intersec.neq(0)
  inter_final = intersec.mask(inter_mask)

  union_mask = union.neq(0)
  union_final = union.mask(union_mask)

  inter_pc = inter_final.reduceRegion(
      reducer=ee.Reducer.count(),
      crs='EPSG:32603',
      scale=4
  )

  union_pc = union_final.reduceRegion(
      reducer=ee.Reducer.count(),
      crs = 'EPSG:32603',
      scale=4
  )

  inter_count = inter_pc.getInfo()["water"]
  union_count = union_pc.getInfo()["water"]

  iou = inter_count / union_count

  ret_string = '{4} -- IoU: {0:0.3f}, Accuracy: {1:0.3f}, Precision: {2:0.3f}, Recall: {3:0.3f}'.format(iou, acc, prec, rec, desc)
  print(ret_string)

  return #ret_string#iou, acc, prec, rec
```
